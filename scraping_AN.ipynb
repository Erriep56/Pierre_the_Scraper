```python
# Paramètres
nom_projet = "lyautey"
base_lien = f"https://www.siv.archives-nationales.culture.gouv.fr/mm/media/download/"
chemin_csv = f"{nom_projet}.csv"

album = 1
premiere_image = 2545
derniere_image = 2871
total_image = derniere_image - premiere_image + 1
```

```python
import os
import csv
import requests
import time

def extraire_numero(nom_image):
    return int(nom_image.split("_")[2])
# seulement si pas de csv à lire

if os.path.exists(chemin_csv):
    with open(chemin_csv, newline='', encoding='utf-8') as fichier_csv:
        lecteur = csv.DictReader(fichier_csv, delimiter=';')
        
        for ligne in lecteur:
            album = ligne['album']
            start_image = ligne['start']
            end_image = ligne['end']

            debut = extraire_numero(start_image)
            fin = extraire_numero(end_image)
            total_image = fin - debut + 1
            numero_telechargement = 0

            # Création du dossier de destination
            images_folder = f"images/{nom_projet}/album_{album}"
            os.makedirs(images_folder, exist_ok=True)

            for i in range(debut, fin + 1):
                numero_telechargement += 1
                numero = str(i).zfill(4)
                nom_image = f"FRAN_0252_{numero}_L-medium.jpg"
                url = base_lien + nom_image
                print(f"[Album {album}] {numero_telechargement}/{total_image} : {url}")

                chemin_fichier = os.path.join(images_folder, nom_image)
                if os.path.exists(chemin_fichier):
                    print(f"✓ Image déjà présente : {chemin_fichier}")
                    continue

                try:
                    response = requests.get(url)
                    if response.status_code == 200:
                        with open(chemin_fichier, 'wb') as f:
                            f.write(response.content)
                        print(f"→ Enregistré : {chemin_fichier}")
                    else:
                        print(f"✗ Erreur {response.status_code} pour {url}")
                except Exception as e:
                    print(f"✗ Exception pour {url} : {e}")
                # time.sleep(2)

else:
    numero_telechargement = 0
    images_folder = f"images/{nom_projet}/album_{album}"
    os.makedirs(images_folder, exist_ok=True)
    for i in range(premiere_image, derniere_image + 1):
        numero_telechargement += 1
        numero = str(i).zfill(5)  # Ex : "02545"
        suppl_lien = f"FRAN_0348_{numero}_L-medium.jpg"
        url = base_lien + suppl_lien
        print(f"{numero_telechargement}/{total_image} Téléchargement de : {url}")

        chemin_fichier = os.path.join(images_folder, f"FRAN_0348_{i+1}.jpg")
        if os.path.exists(chemin_fichier):
            print(f"l'image {chemin_fichier} existe déjà")
            continue
        try:
            response = requests.get(url)
            if response.status_code == 200:
                with open(chemin_fichier, 'wb') as f:
                    f.write(response.content)
                print(f"→ Enregistré sous : {chemin_fichier}")
            else:
                print(f"✗ Erreur {response.status_code} pour {url}")
        except Exception as e:
            print(f"✗ Exception pour {url} : {e}")
        # time.sleep(2)
```
